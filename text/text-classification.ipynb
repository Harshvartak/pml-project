```python
# Step 1: Exploratory Data Analysis (EDA) for Text Dataset

import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import nltk
import re
from wordcloud import WordCloud

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Load the dataset
X_train_text = pickle.load(open(local_text_dataset, 'rb'))

# Convert to DataFrame for easier manipulation
df = pd.DataFrame({'document': X_train_text})

# Preview the data
print("Number of documents:", len(df))
print("\nSample documents:")
print(df['document'].head(2))  # Show only 2 examples

# Remove empty documents
df['length'] = df['document'].apply(lambda x: len(x.strip()))
df = df[df['length'] > 0]
print("\nNumber of non-empty documents:", len(df))

# Check for duplicates
duplicates = df.duplicated().sum()
print("\nNumber of duplicate documents:", duplicates)

# Document length distribution
df['word_count'] = df['document'].apply(lambda x: len(word_tokenize(x)))
plt.figure(figsize=(10, 5))
sns.histplot(df['word_count'], bins=30, kde=True)
plt.title('Distribution of Word Count per Document')
plt.xlabel('Word Count')
plt.ylabel('Number of Documents')
plt.show()

# Improved preprocessing with Lemmatization using NLTK
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
def preprocess(doc):
    doc = doc.lower()
    doc = re.sub(r"[^a-zA-Z\s]", "", doc)
    tokens = word_tokenize(doc)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and len(word) > 2]
    return tokens

# Flatten all tokens into a list
token_lists = df['document'].apply(preprocess)
all_tokens = [token for sublist in token_lists for token in sublist]

# Show 2 example processed texts
print("\nExample preprocessed documents:")
print(token_lists.head(2))

# Plot top 30 most frequent words
common_words = Counter(all_tokens).most_common(30)
words, freqs = zip(*common_words)
plt.figure(figsize=(12, 6))
sns.barplot(x=list(freqs), y=list(words))
plt.title("Top 30 Most Common Words")
plt.xlabel("Frequency")
plt.ylabel("Word")
plt.show()

# WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_tokens))
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Frequent Words in Corpus")
plt.show()


# Step 2: Text Preprocessing and Vectorization

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
import numpy as np

# Join cleaned tokens back to text for TF-IDF
df['clean_text'] = token_lists.apply(lambda tokens: ' '.join(tokens))

# Show 2 examples of cleaned text
print("\nExample cleaned texts for TF-IDF:")
print(df['clean_text'].head(2))

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_df=0.8, min_df=5, max_features=10000)
X_tfidf = vectorizer.fit_transform(df['clean_text'])

print("\nTF-IDF matrix shape:", X_tfidf.shape)

# Dimensionality Reduction with Truncated SVD
svd = TruncatedSVD(n_components=100, random_state=42)
normalizer = Normalizer(copy=False)
lsa = make_pipeline(svd, normalizer)

X_lsa = lsa.fit_transform(X_tfidf)
print("LSA-reduced feature matrix shape:", X_lsa.shape)

# Explained variance
explained_variance = svd.explained_variance_ratio_.sum()
print("\nTotal explained variance by 100 components:", round(explained_variance, 4))

# Plot explained variance by components
plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(svd.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance by LSA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()


# Step 3: Clustering and Self-Supervised Labeling

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

# Apply KMeans clustering
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
y_kmeans = kmeans.fit_predict(X_lsa)

# Store cluster labels in the DataFrame for interpretation and training
df['pseudo_label'] = y_kmeans

# Apply Agglomerative Clustering
auto_agglom = AgglomerativeClustering(n_clusters=n_clusters)
y_agglom = auto_agglom.fit_predict(X_lsa)

# Compute clustering metrics
print("\nKMeans Silhouette Score:", silhouette_score(X_lsa, y_kmeans))
print("KMeans Davies-Bouldin Index:", davies_bouldin_score(X_lsa, y_kmeans))
print("Agglomerative Silhouette Score:", silhouette_score(X_lsa, y_agglom))
print("Agglomerative Davies-Bouldin Index:", davies_bouldin_score(X_lsa, y_agglom))

# Visualize clusters using PCA
pca_2d = PCA(n_components=2)
x_pca = pca_2d.fit_transform(X_lsa)

plt.figure(figsize=(10, 5))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y_kmeans, cmap='Set1', s=10)
plt.title("KMeans Clusters Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()

# Visualize clusters using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
x_tsne = tsne.fit_transform(X_lsa)

plt.figure(figsize=(10, 5))
plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c=y_kmeans, cmap='Set2', s=10)
plt.title("KMeans Clusters Visualized with t-SNE")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()

print("Top Terms Per Cluster:")
from sklearn.feature_extraction.text import TfidfVectorizer
X_tfidf_array = X_tfidf.toarray()
feature_names = vectorizer.get_feature_names_out()

cluster_keywords = {}
for cluster in range(n_clusters):
    indices = df[df['pseudo_label'] == cluster].index
    cluster_vecs = X_tfidf_array[indices]
    avg_tfidf = np.mean(cluster_vecs, axis=0)
    top_indices = avg_tfidf.argsort()[-10:][::-1]
    top_words = [feature_names[i] for i in top_indices]
    cluster_keywords[cluster] = top_words
    readable_label = ' / '.join(top_words[:3])  # Use top 3 words as label
    print(f"Cluster {cluster} (aka '{readable_label}') Top Words:", ', '.join(top_words))
    df.loc[df['pseudo_label'] == cluster, 'cluster_label'] = readable_label

# Assign single-word cluster names dynamically based on top TF-IDF keywords

from collections import Counter

def get_cluster_topic_name(top_words):
    # Select the most frequent unique word that appears most across top keywords
    return top_words[0] if top_words else 'Unknown'

cluster_name_map = {}

for cluster, keywords in cluster_keywords.items():
    topic_name = get_cluster_topic_name(keywords)
    cluster_name_map[cluster] = topic_name

# Map pseudo_label to human-readable cluster_name
# Adds new column 'cluster_name' that maps KMeans numeric cluster to top TF-IDF keyword
df['cluster_name'] = df['pseudo_label'].map(cluster_name_map)

print("Assigned Cluster Names:")
for cluster, name in cluster_name_map.items():
    print(f"Cluster {cluster}: {name}")

# Show 2 sample documents per cluster
print("Sample Documents Per Cluster:")
for cluster in range(n_clusters):
    print(f"--- Cluster {cluster} ---")
    samples = df[df['pseudo_label'] == cluster]['document'].sample(n=2, random_state=cluster).values
    for doc in samples:
        print(f"• {doc[:200]}...")

# Visualize clusters using UMAP
import umap

umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
x_umap = umap_model.fit_transform(X_lsa)

plt.figure(figsize=(10, 5))
plt.scatter(x_umap[:, 0], x_umap[:, 1], c=y_kmeans, cmap='tab10', s=10)
plt.title("KMeans Clusters Visualized with UMAP")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()

# Step 4: Supervised Classification Using Pseudo-Labels

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, f1_score
import seaborn as sns

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_lsa, df['pseudo_label'], test_size=0.2, random_state=42, stratify=df['pseudo_label'])

# Initialize classifiers
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42)
}

# Evaluate and compare models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"{name} Classification Report:")
    print(classification_report(y_test, y_pred, target_names=[cluster_name_map[i] for i in sorted(set(y_test))]))
    print("Confusion Matrix:")

from sklearn.utils.multiclass import unique_labels
labels = unique_labels(y_test, y_pred)
label_names = [cluster_name_map[label] for label in labels]  # Cluster name mapping for confusion matrix readability
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=label_names, yticklabels=label_names)
plt.title(f'{name} Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Analyze class distribution
plt.figure(figsize=(6,4))
df['cluster_name'].value_counts().reindex(list(cluster_name_map.values())).plot(kind='bar', color='skyblue')
plt.title('Cluster Name Distribution')
plt.xlabel('Cluster Name')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# Apply SMOTE to handle imbalance
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
print("Balanced label distribution after SMOTE:")
print(pd.Series(y_train_bal).value_counts())

# Visualize class distribution after SMOTE
plt.figure(figsize=(6,4))
pd.Series(y_train_bal).map(cluster_name_map).value_counts().reindex(list(cluster_name_map.values())).plot(kind='bar', color='lightgreen')
plt.title('Balanced Cluster Name Distribution After SMOTE')
plt.xlabel('Cluster Name')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# Hyperparameter tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV

# Tune Random Forest
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
}
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=3, scoring='f1_macro')
rf_grid.fit(X_train_bal, y_train_bal)
best_rf = rf_grid.best_estimator_

# Tune Logistic Regression
param_grid_lr = {
    'C': [0.1, 1.0, 10.0]
}
lr_grid = GridSearchCV(LogisticRegression(max_iter=1000, random_state=42), param_grid_lr, cv=3, scoring='f1_macro')
lr_grid.fit(X_train_bal, y_train_bal)
best_lr = lr_grid.best_estimator_

# Tune SVM
param_grid_svm = {
    'C': [0.1, 1.0, 10.0],
    'kernel': ['linear', 'rbf']
}
svm_grid = GridSearchCV(SVC(random_state=42), param_grid_svm, cv=3, scoring='f1_macro')
svm_grid.fit(X_train_bal, y_train_bal)
best_svm = svm_grid.best_estimator_

# Evaluate best models
# The classifiers below are trained using SMOTE-balanced data (X_train_bal, y_train_bal)
# to ensure fair representation of all cluster classes during learning.
# Assess performance of the best-tuned classifiers using held-out test data.
# Metrics and confusion matrices provide insight into how well each model generalizes.

# Visual Comparison: Baseline vs SMOTE
baseline_scores = {
    'Random Forest': f1_score(y_test, models['Random Forest'].predict(X_test), average='macro'),
    'Logistic Regression': f1_score(y_test, models['Logistic Regression'].predict(X_test), average='macro'),
    'SVM': f1_score(y_test, models['SVM'].predict(X_test), average='macro')
}

tuned_scores = {
    'Random Forest': f1_score(y_test, best_rf.predict(X_test), average='macro'),
    'Logistic Regression': f1_score(y_test, best_lr.predict(X_test), average='macro'),
    'SVM': f1_score(y_test, best_svm.predict(X_test), average='macro')
}

# Create bar chart
x_labels = list(baseline_scores.keys())
x = np.arange(len(x_labels))
bar_width = 0.35

plt.figure(figsize=(10, 5))
plt.bar(x, baseline_scores.values(), width=bar_width, label='Baseline (Imbalanced)', color='salmon')
plt.bar(x + bar_width, tuned_scores.values(), width=bar_width, label='After SMOTE', color='seagreen')

plt.xlabel('Model')
plt.ylabel('Macro F1 Score')
plt.title('Comparison of Classifier Performance: Baseline vs SMOTE')
plt.xticks(x + bar_width / 2, x_labels)
plt.ylim(0.6, 1.0)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Evaluate best models

import numpy as np
importances = best_rf.feature_importances_
feature_names = vectorizer.get_feature_names_out()
indices = np.argsort(importances)[-20:][::-1]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[indices], y=np.array(feature_names)[indices])
plt.title("Top 20 Important Features from Random Forest")
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.show()

# Compare with BERT/SBERT embeddings
from sentence_transformers import SentenceTransformer
from sklearn.metrics import silhouette_score

# Use SBERT to generate embeddings
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
X_sbert = sbert_model.encode(df['clean_text'], show_progress_bar=True)

# Re-cluster using KMeans on SBERT
kmeans_sbert = KMeans(n_clusters=5, random_state=42)
y_sbert = kmeans_sbert.fit_predict(X_sbert)

# Evaluate clustering quality
print("SBERT Clustering Silhouette Score:", silhouette_score(X_sbert, y_sbert))

# Visualize SBERT clusters using UMAP
x_sbert_umap = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42).fit_transform(X_sbert)
plt.figure(figsize=(10, 5))
plt.scatter(x_sbert_umap[:, 0], x_sbert_umap[:, 1], c=y_sbert, cmap='tab10', s=10)
plt.title("SBERT Clusters Visualized with UMAP")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()

# Compare SBERT vs TF-IDF: Classification Accuracy
# Train classifier on SBERT features using best Logistic Regression for fair comparison
X_train_sbert, X_test_sbert, y_train_sbert, y_test_sbert = train_test_split(X_sbert, df['pseudo_label'], test_size=0.2, random_state=42, stratify=df['pseudo_label'])

best_lr_sbert = LogisticRegression(C=best_lr.C, max_iter=1000, random_state=42)
best_lr_sbert.fit(X_train_sbert, y_train_sbert)
y_pred_sbert = best_lr_sbert.predict(X_test_sbert)

print("Logistic Regression on SBERT Embeddings:")
print(classification_report(y_test_sbert, y_pred_sbert, target_names=[cluster_name_map[i] for i in sorted(set(y_test_sbert))]))
sns.heatmap(confusion_matrix(y_test_sbert, y_pred_sbert), annot=True, fmt='d', cmap='Purples')
plt.title('Confusion Matrix - SBERT')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Consolidated performance summary across methods
print("--- Performance Summary ---")
print("TF-IDF + Tuned Logistic Regression")
print(classification_report(y_test, best_lr.predict(X_test), target_names=[cluster_name_map[i] for i in sorted(set(y_test))]))
print("SBERT + Logistic Regression")
print(classification_report(y_test_sbert, y_pred_sbert, target_names=[cluster_name_map[i] for i in sorted(set(y_test_sbert))]))

# Visualize BERT Clusters and Compare to SBERT
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Generate BERT embeddings (CLS token mean pooling)
def get_bert_embeddings(texts):
    bert_model.eval()
    embeddings = []
    with torch.no_grad():
        for text in texts:
            inputs = bert_tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
            outputs = bert_model(**inputs)
            cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze().numpy()
            embeddings.append(cls_embedding)
    return np.vstack(embeddings)

print("Generating BERT embeddings...")
X_bert = get_bert_embeddings(df['clean_text'])

# KMeans clustering on BERT
kmeans_bert = KMeans(n_clusters=5, random_state=42)
y_bert = kmeans_bert.fit_predict(X_bert)

# Silhouette Score for BERT
print("BERT Clustering Silhouette Score:", silhouette_score(X_bert, y_bert))

# Visualize BERT clusters with UMAP
x_bert_umap = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42).fit_transform(X_bert)
plt.figure(figsize=(10, 5))
plt.scatter(x_bert_umap[:, 0], x_bert_umap[:, 1], c=y_bert, cmap='Spectral', s=10)
plt.title("BERT Clusters Visualized with UMAP")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()

# Compare BERT vs SBERT with classification
X_train_bert, X_test_bert, y_train_bert, y_test_bert = train_test_split(X_bert, df['pseudo_label'], test_size=0.2, random_state=42, stratify=df['pseudo_label'])

best_lr_bert = LogisticRegression(C=best_lr.C, max_iter=1000, random_state=42)
best_lr_bert.fit(X_train_bert, y_train_bert)
y_pred_bert = best_lr_bert.predict(X_test_bert)

print("Logistic Regression on BERT Embeddings:")
print(classification_report(y_test_bert, y_pred_bert, target_names=[cluster_name_map[i] for i in sorted(set(y_test_bert))]))
sns.heatmap(confusion_matrix(y_test_bert, y_pred_bert), annot=True, fmt='d', cmap='Oranges')
plt.title('Confusion Matrix - BERT')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Final Consolidated Summary

# Compute macro F1 values before identifying best model
from sklearn.metrics import accuracy_score, f1_score

model_names = [
    'TF-IDF + Logistic Regression',
    'TF-IDF + Random Forest',
    'TF-IDF + SVM',
    'SBERT + Logistic Regression',
    'BERT + Logistic Regression'
]

accuracy_values = [
    accuracy_score(y_test, best_lr.predict(X_test)),
    accuracy_score(y_test, best_rf.predict(X_test)),
    accuracy_score(y_test, best_svm.predict(X_test)),
    accuracy_score(y_test_sbert, y_pred_sbert),
    accuracy_score(y_test_bert, y_pred_bert)
]

macro_f1_values = [
    f1_score(y_test, best_lr.predict(X_test), average='macro'),
    f1_score(y_test, best_rf.predict(X_test), average='macro'),
    f1_score(y_test, best_svm.predict(X_test), average='macro'),
    f1_score(y_test_sbert, y_pred_sbert, average='macro'),
    f1_score(y_test_bert, y_pred_bert, average='macro')
]

# Identify the best model dynamically based on macro F1 score
macro_f1_map = {
    'TF-IDF + Logistic Regression': best_lr,
    'TF-IDF + Random Forest': best_rf,
    'TF-IDF + SVM': best_svm,
    'SBERT + Logistic Regression': best_lr_sbert,
    'BERT + Logistic Regression': best_lr_bert
}

best_index = np.argmax(macro_f1_values)
best_model_name = model_names[best_index]
best_model_object = macro_f1_map[best_model_name]

print(f"Best performing model based on Macro F1: {best_model_name}")

# Save the best model pipeline

import joblib
joblib.dump(best_model_object, 'final_best_model.pkl')
joblib.dump(vectorizer, 'final_vectorizer.pkl')
joblib.dump(svd, 'final_svd.pkl')

print("--- Final Model Comparison ---")
print("TF-IDF + Logistic Regression")
print(classification_report(y_test, best_lr.predict(X_test), target_names=[cluster_name_map[i] for i in sorted(set(y_test))]))
print("SBERT + Logistic Regression")
print(classification_report(y_test_sbert, y_pred_sbert, target_names=[cluster_name_map[i] for i in sorted(set(y_test_sbert))]))
print("BERT + Logistic Regression (Cluster Names)")
print(classification_report(y_test_bert, y_pred_bert, target_names=[cluster_name_map[i] for i in sorted(set(y_test_bert))]))

# Visualize accuracy and macro F1 comparison across models
models = [
    'TF-IDF + Logistic Regression',
    'TF-IDF + Random Forest',
    'TF-IDF + SVM',
    'SBERT + Logistic Regression',
    'BERT + Logistic Regression'
]

from sklearn.metrics import accuracy_score, f1_score

# Dynamically compute accuracies and macro F1 scores
model_names = [
    'TF-IDF + Logistic Regression',
    'TF-IDF + Random Forest',
    'TF-IDF + SVM',
    'SBERT + Logistic Regression',
    'BERT + Logistic Regression'
]

accuracy_values = [
    accuracy_score(y_test, best_lr.predict(X_test)),
    accuracy_score(y_test, best_rf.predict(X_test)),
    accuracy_score(y_test, best_svm.predict(X_test)),
    accuracy_score(y_test_sbert, y_pred_sbert),
    accuracy_score(y_test_bert, y_pred_bert)
]

macro_f1_values = [
    f1_score(y_test, best_lr.predict(X_test), average='macro'),
    f1_score(y_test, best_rf.predict(X_test), average='macro'),
    f1_score(y_test, best_svm.predict(X_test), average='macro'),
    f1_score(y_test_sbert, y_pred_sbert, average='macro'),
    f1_score(y_test_bert, y_pred_bert, average='macro')
]

x = range(len(model_names))
bar_width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x, accuracy_values, width=bar_width, label='Accuracy', color='steelblue')
plt.bar([i + bar_width for i in x], macro_f1_values, width=bar_width, label='Macro F1', color='orange')

plt.xlabel('Model')
plt.ylabel('Score')
plt.title('Model Comparison: Accuracy vs Macro F1')
plt.xticks([i + bar_width / 2 for i in x], model_names, rotation=30, ha='right')
plt.ylim(0.7, 1.0)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
